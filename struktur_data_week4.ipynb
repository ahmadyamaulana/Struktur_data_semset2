```python
class antriantiket:
    def __init__(Self, max_capacity=10):
        Self.antrian = []
        Self.max_capacity = max_capacity

    def tambah_antrian(Self, nama):
        if Self.size() < Self.max_capacity:
            Self.antrian.append(nama)
            print(f"{nama} ditambahkan ke antrian.")
        else:
            print("Antrian penuh. Mohon tunggu.")

    def tambah_antrian_vip(Self, nama):
        Self.antrian.insert(0, nama)
        print(f"{nama} (VIP) ditambahkan ke antrian.")

    def hapus_antrian(Self):
        if not Self.is_empty():
            nama = Self.antrian.pop(0)
            print(f"{nama} dihapus dari antrian.")
            return nama
        else:
            print("Antrian kosong.")
            return None

    def front(self):
        return self.antrian[0] if not self.is_empty() else None

    def rear(self):
        return self.antrian[-1] if not self.is_empty() else None

    def is_empty(Self):
        return len(Self.antrian) == 0

    def size(Self):
        return len(Self.antrian)

queue = antriantiket()
queue.tambah_antrian("IL_67")
queue.tambah_antrian("GI_68")
queue.tambah_antrian("AH_69")
print("Antrian setelah ditambahkan:")
print(queue.antrian)
print("Front Antrian:", queue.front())
print("Rear Antrian:", queue.rear())
print("Ukuran Antrian:", queue.size())
```

```python
class Queue:
    def __init__(self, capacity=10):
        self.queue = []
        self.capacity = capacity
        self.size = 0

    def get_size(self):
        return self.size

    def enqueue(self, ticket_id):
        if self.size == self.capacity:
            print("Antrian penuh, mohon tunggu.")
            return False
        else:
            self.queue.append(ticket_id)
            self.size += 1
            print(f"Pelanggan dengan tiket {ticket_id} ditambahkan ke antrian.")
            return True

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        else:
            ticket_id = self.queue.pop(0)
            self.size -= 1
            print(f"Pelanggan dengan tiket {ticket_id} dilayani.")
            return ticket_id

    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        else:
            return self.queue[0]

    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        else:
            return self.queue[-1]

    def is_empty(self):
        return self.size == 0

    def size(self):
        return self.size

    def is_vip(self, ticket_id):
        # Cek apakah kode pelanggan mengandung angka 1
        return '1' in ticket_id

# Contoh penggunaan
queue = Queue(capacity=10)

# Membuat kode pelanggan dengan format "AD" + "37"
kode_1 = "AD3701"
kode_2 = "AD3702"
kode_3 = "AD3703"

# Menambahkan pelanggan ke antrian
queue.enqueue(kode_1)
queue.enqueue(kode_2)
queue.enqueue(kode_3)

# Cek pelanggan VIP
print(f"Apakah {kode_1} adalah pelanggan VIP? {queue.is_vip(kode_1)}")
print(f"Apakah {kode_2} adalah pelanggan VIP? {queue.is_vip(kode_2)}")
print(f"Apakah {kode_3} adalah pelanggan VIP? {queue.is_vip(kode_3)}")

# Melihat pelanggan pertama dan terakhir di antrian
print(f"Pelanggan pertama dalam antrian: {queue.front()}")
print(f"Pelanggan terakhir dalam antrian: {queue.rear()}")

# Melayani pelanggan
queue.dequeue()

# Cek ukuran antrian
print(f"Ukuran antrian saat ini: {queue.get_size()}")
```

# **Tugas 2**

```python
class PhotocopyQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            self.queue.insert(0, (order_id, num_pages))
            print(f"Pesanan {order_id} dengan {num_pages} halaman langsung diproses!")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} dengan {num_pages} halaman ditambahkan ke antrian.")

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong! Tidak ada pesanan yang bisa diproses.")
            return None

        order = self.queue.pop(0)
        order_id, num_pages = order

        if num_pages > 100:
            remaining_pages = num_pages - 100
            self.queue.append((order_id, remaining_pages))
            print(f"Pesanan {order_id} diproses 100 halaman, sisanya {remaining_pages} halaman masuk kembali ke antrian.")
        else:
            print(f"Pesanan {order_id} dengan {num_pages} halaman telah diproses.")
        return order

    def front(self):
        return "Antrian kosong" if self.is_empty() else self.queue[0]

    def rear(self):
        return "Antrian kosong" if self.is_empty() else self.queue[-1]

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# IMPLEMENTASI
queue = PhotocopyQueue()
queue.enqueue("AHM01", 230)
queue.enqueue("ILH02", 10)
queue.enqueue("GHI03", 6)
queue.enqueue("NOP04", 190)

print("")
print("Pesanan pertama dalam antrian:", queue.front())
print("Pesanan terakhir dalam antrian:", queue.rear())
print("Total pesanan dalam antrian:", queue.size())

print("")
queue.dequeue()
queue.dequeue()
queue.dequeue()

print("")
print("Total pesanan dalam antrian setelah pemrosesan:", queue.size())
```

```python
class AntrianParkir:
    def __init__(self, kapasitas=15):
        self.antrian = []
        self.kapasitas = kapasitas

    def enqueue(self, vehicle_number, is_electric=False):
        if self.size() >= self.kapasitas:
            print(f"Parkiran penuh! Kendaraan {vehicle_number} harus menunggu.")
            return False

        if is_electric:
            self.antrian.insert(0, vehicle_number)  # Prioritaskan kendaraan listrik di depan
            print(f"Kendaraan listrik {vehicle_number} ditambahkan ke antrian dengan prioritas.")
        else:
            self.antrian.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} ditambahkan ke antrian.")
        return True

    def dequeue(self):
        """Mengeluarkan kendaraan yang lebih dulu datang dari parkiran."""
        if self.is_empty():
            print("Antrian parkir kosong.")
            return None
        vehicle_number = self.antrian.pop(0)
        print(f"Kendaraan {vehicle_number} keluar dari parkiran.")
        return vehicle_number

    def front(self):
        if self.is_empty():
            print("Antrian parkir kosong.")
            return None
        return self.antrian[0]

    def rear(self):
        if self.is_empty():
            print("Antrian parkir kosong.")
            return None
        return self.antrian[-1]

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

# Contoh Penggunaan
parkir = AntrianParkir(kapasitas=15)

# Menambahkan kendaraan
for i in range(1, 17):  # Menambahkan lebih dari kapasitas
    parkir.enqueue(f"TE11{i:02d}")

# Menampilkan informasi antrian
print("Kendaraan pertama dalam antrian:", parkir.front())
print("Kendaraan terakhir dalam antrian:", parkir.rear())
print("Jumlah kendaraan dalam antrian:", parkir.size())

# Mengeluarkan kendaraan dari antrian
parkir.dequeue()

# Mencoba menambahkan kendaraan setelah ada yang keluar
parkir.enqueue("TE1200")

# Memeriksa apakah antrian kosong
print("Apakah antrian kosong?", parkir.is_empty())
```

# **Tugas 4**

```python
class AntrianSertifikat:
    def __init__(self):
        self.antrian = []
        self.antrian_reguler = []
        self.antrian_prestasi = []
        self.batas_antrian = 20
        self.nomor_antrian = 1  # Inisialisasi nomor antrian

    def enqueue(self, student_id, prestasi=False):
        """Menambahkan mahasiswa ke dalam antrian."""
        nomor_antrian = f"S37-{self.nomor_antrian:03d}"  # Format nomor antrian
        self.nomor_antrian += 1

        if prestasi:
            print(f"Mahasiswa {student_id} ({nomor_antrian}) langsung dilayani karena prestasi!")
            return  # Langsung dilayani, tidak perlu masuk antrian

        if len(self.antrian) >= self.batas_antrian:
            if prestasi:
                self.antrian_prestasi.append((student_id, nomor_antrian))
                print(f"Mahasiswa {student_id} ({nomor_antrian}) masuk antrian prestasi.")
            else:
                self.antrian_reguler.append((student_id, nomor_antrian))
                print(f"Mahasiswa {student_id} ({nomor_antrian}) masuk antrian reguler.")
        else:
            self.antrian.append((student_id, nomor_antrian))
            print(f"Mahasiswa {student_id} ({nomor_antrian}) masuk antrian utama.")

    def dequeue(self):
        """Melayani mahasiswa dari antrian depan."""
        if self.is_empty():
            print("Antrian kosong.")
            return None

        if self.antrian:
            student_id, nomor_antrian = self.antrian.pop(0)
            print(f"Melayani mahasiswa {student_id} ({nomor_antrian}) dari antrian utama.")
            return student_id, nomor_antrian
        elif self.antrian_reguler:
            student_id, nomor_antrian = self.antrian_reguler.pop(0)
            print(f"Melayani mahasiswa {student_id} ({nomor_antrian}) dari antrian reguler.")
            return student_id, nomor_antrian
        else:
            student_id, nomor_antrian = self.antrian_prestasi.pop(0)
            print(f"Melayani mahasiswa {student_id} ({nomor_antrian}) dari antrian prestasi.")
            return student_id, nomor_antrian

    def front(self):
        """Melihat mahasiswa pertama dalam antrian."""
        if self.is_empty():
            print("Antrian kosong.")
            return None

        if self.antrian:
            student_id, nomor_antrian = self.antrian[0]
            print(f"Mahasiswa pertama dalam antrian utama: {student_id} ({nomor_antrian})")
            return student_id, nomor_antrian
        elif self.antrian_reguler:
            student_id, nomor_antrian = self.antrian_reguler[0]
            print(f"Mahasiswa pertama dalam antrian reguler: {student_id} ({nomor_antrian})")
            return student_id, nomor_antrian
        else:
            student_id, nomor_antrian = self.antrian_prestasi[0]
            print(f"Mahasiswa pertama dalam antrian prestasi: {student_id} ({nomor_antrian})")
            return student_id, nomor_antrian

    def rear(self):
        """Melihat mahasiswa terakhir dalam antrian."""
        if self.is_empty():
            print("Antrian kosong.")
            return None

        if self.antrian:
            student_id, nomor_antrian = self.antrian[-1]
            print(f"Mahasiswa terakhir dalam antrian utama: {student_id} ({nomor_antrian})")
            return student_id, nomor_antrian
        elif self.antrian_reguler:
            student_id, nomor_antrian = self.antrian_reguler[-1]
            print(f"Mahasiswa terakhir dalam antrian reguler: {student_id} ({nomor_antrian})")
            return student_id, nomor_antrian
        else:
            student_id, nomor_antrian = self.antrian_prestasi[-1]
            print(f"Mahasiswa terakhir dalam antrian prestasi: {student_id} ({nomor_antrian})")
            return student_id, nomor_antrian

    def is_empty(self):
        """Memeriksa apakah antrian kosong."""
        return len(self.antrian) == 0 and len(self.antrian_reguler) == 0 and len(self.antrian_prestasi) == 0

    def size(self):
        """Menghitung jumlah mahasiswa dalam antrian."""
        total_size = len(self.antrian) + len(self.antrian_reguler) + len(self.antrian_prestasi)
        print(f"Jumlah mahasiswa dalam antrian: {total_size}")
        return total_size

# Contoh Penggunaan
antrian = AntrianSertifikat()

# Mahasiswa yang pernah ikut kompetisi (langsung dilayani)
antrian.enqueue("2241720001", prestasi=True)

# Mahasiswa reguler
antrian.enqueue("2241720002")
antrian.enqueue("2241720003")
antrian.enqueue("2241720004")

# Tambahkan lebih banyak mahasiswa untuk memenuhi batas antrian
for i in range(17):
    antrian.enqueue(f"224172000{i+5}")

# Setelah antrian penuh, mahasiswa dengan prestasi masuk ke antrian prestasi
antrian.enqueue("2241720022", prestasi=True)

# Mahasiswa reguler setelah antrian penuh
antrian.enqueue("2241720023")
antrian.enqueue("2241720024")

antrian.size()
antrian.front()
antrian.rear()

# Melayani beberapa mahasiswa
antrian.dequeue()
antrian.dequeue()

antrian.size()
```

```python
#definisi node dalam doubly linked list
class pageNode:
  def __init__(self,page_number):
    self.page_number = page_number #menyimpan nomor halaman
    self.next = None #pointer ke halaman berikutnya
    self.prev = None #pointer ke halaman sebelumnya

#definisi doubly linked list untuk riwayat pembayaran
class BrowserHistory:
  def __init__(self):
    self.current = None #pointer ke halaman yang sedang terbuka

  #menambahkan halaman baru ke dalam riwayat
  def visit_page(self, url):
    # Corrected the class name from PageNone to pageNode
    new_page = pageNode(url)
    if not self.current:
      self.current = new_page #jika riwayat kosong halaman ini menjadi halaman pertama
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page #Pindahkan posisi ke halaman baru

  #menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print("\n Riwayat kosong.")

    print("\n Riwayat Peramban web:")
    # Changed tempt to temp for consistency
    temp = self.current
    while temp.prev: #kembali ke halaman pertama
        temp = temp.prev

    count = 1
    while temp:
        # Added url attribute to pageNode to store URL
        print(f"{count}. {temp.page_number}")
        temp = temp.next
        count += 1

  #Navigasi mundur ke halaman sebelumnya
  def go_back(self):
      if self.current and self.current.prev:
        self.current = self.current.prev
        # Accessing page_number attribute instead of url
        print(f"\n Anda kembali ke: {self.current.page_number}")
      else:
        print("\n Tidak bisa kembali, ini adalah halaman pertama.")

  #Navigasi maju ke halaman berikutnya
  def go_forward(self):
      if self.current and self.current.next:
        self.current = self.current.next
        # Accessing page_number attribute instead of url
        print(f"\n Anda maju ke: {self.current.page_number}")
      else:
        print("\n Tidak bisa maju, ini adalah halaman terakhir.")

#======== Program Utama ================#

#membuat objek riwayat peramban
history = BrowserHistory()

#meminta pengguna memasukan jumlah halaman yang dikunjungi
num_pages = int(input("\nMasukan jumlah halaman yang dikunjungi:"))

#loop untuk memasukkan URL halaman ke dalam Doubly Linked list
for i in range(1,num_pages + 1):
  url = input(f"Masukan URL halaman ke-{i}:")
  history.visit_page(url)

#menampilkan riwayat kunjungan
history.show_history()

#menu navigasi pengguna
while True:
  # Changed commad to command
  command = input("\n Tekan 'b' untuk kembali, 'f' untuk maju, atau 'x' untuk keluar: ")

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n keluar dari riwayat pramban.")
    break
  else:
    print("\n Perintah tidak dikenal")
```

```python
#definisi Node dalam single linked list
class Node:
    def __init__(self,data):
      self.data = data #menyimpan data
      self.next = None #pointer ke node setelahnya

# definisi single linked list dengan berbagai operasi
class LinkedList:
    def __init__(self):
        self.head = None #awal, daftar kosong

    #menambahkan elemen di awal
    def add_at_beginning(self,data):
      new_node = Node(data)
      new_node.next = self.head
      self.head = new_node

    def add_at_end(self,data):
      new_node = Node(data)
      if not self.head:
        self.head = new_node
        return
      temp = self.head
      while temp.next:
        temp = temp.next
      temp.next = new_node

    #menambahkan elemen di posisi tertentu
    def add_at_position(self, positon, data):
        if position < 0:
            print("posisi tidak valid")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    #menghapus elemen diawal
    def delete_at_beginning(self):
        if not self.head:
          print("Daftar kosong, tidak adayang bisa dihapus!")
          return
        self.head = self.head.next

    #menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    #menghapus elemen di posisi teretentu
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.delete_at_beginning()
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print("posisi tidak ditemukan")
                return
            temp = temp.next
        temp.next = temp.next.next

    #mencari elemen dalam linked list
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar.")
                return
            temp = temp.next
        print(f"Elemen {key} tidak ditemukan dalam daftar.")

    #membalik (reverse) linked list
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    #menampilkan liked list
    def display(slef):
        if not self.head:
          print("Daftar kosong")
          return
        temp = self.head
        while temp:
          print(temp.data, end="->")
          temp = temp.next
        print("None")

#============== Program Utama ============#

#membuat objek linked list
ll = LinkedList()

#meminta pengguna memasukan elemen awal
num_elements = int(input("\nMasukan jumlah awal element dalam draft:"))
for i in range(num_elements):
  data = int(input(f"Masukan data ke-{i+1}:"))
  ll.add_at_end(data)

#menu oprasi linked list
while True:
  print("\nPilih Operasi:")
  print("1. Tambah element Awal")
  print("2. Tambah element akhir")
  print("3. Tambah element di posisi tertentu")
  print("4. Hapus element Awal")
  print("5. Hapus element Awal")
  print("6. Hapus element di posisi tertentu")
  print("7. cari element dalam daftar")
  print("8. balikan daftar")
  print("9. Tampilkan daftar")
  print("0. keluar")

  choice = int(input("\nmasukan pilihan:"))

  if choice == 1:
    data = int(input("Masukan data untuk ditambahkan diawal:"))
    ll.add_at_beginning(data)
  elif choice == 2:
    data = int(input("Masukan data untuk ditambahkan di akhir:"))
    ll.add_at_end(data)
  elif choice == 3:
    position = int(input("Masukan posisi untuk menambahkan data:"))
    data = int(input("Masukan data yang ingin ditambahkan:"))
    ll.add_at_position(position, data)
  elif choice == 4:
    ll.delete_at_beginning()
  elif choice == 5:
    ll.delete_at_end()
  elif choice == 6:
    position = int(input("Masukan posisi untuk menghapus data:"))
    ll.delete_at_position(position)
  elif choice == 7:
    key = int(input("Masukan data yang ingin dicari:"))
    ll.search(key)
  elif choice == 8:
    ll.reverse()
    print("daftar berhasil dibalik")
  elif choice == 9:
    ll.display()
  elif choice == 0:
    print("keluar dari program")
    break
  else:
    print("\n pilihan tidak valid. Silakan coba lagi.")
```
